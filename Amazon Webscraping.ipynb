```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.amazon.in/s?k=oppo+phone&page=2&crid=1OTVS90MF6AVK&qid=1660107552&sprefix=%2Caps%2C611&ref=sr_pg_1'
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
```

```python
page = requests.get(URL, headers=headers)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
mobile = soup.find_all('span',{'class':'a-size-medium a-color-base a-text-normal'})
```

```python
mobile
```

```python
prices = soup.find_all('span',{'class':'a-price-whole'})
```

```python
prices
```

```python
mobile_name = []
for i in range(0,len(mobile)):
    mobile_name.append(mobile[i].get_text())
```

```python
mobile_name
```

```python
mobile_price = []
for i in range(0,len(prices)):
    mobile_price.append(prices[i].get_text())
```

```python
mobile_price
```
